```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
from matplotlib import rc
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] 
```

```python
x = range(0,10)

y = [0,1,2,3,6,2,6,8,2,1]
y_2 = [2,6,8,6,2,3,4,5,6,7]
fig = plt.figure(figsize=(20,8),dpi=80)
_x_ticks = ["{}岁".format(i) for i in x]
plt.xticks(x,_x_ticks, rotation=45)
_y = range(0,9)
_y_ticks = ["{}个".format(i) for i in _y]
plt.yticks(_y,_y_ticks)
plt.xlabel("用户年龄")
plt.ylabel("玩具数量")
plt.plot(x,y,label="自己",color='orange',linestyle="--")
plt.plot(x,y_2,label="同桌",color='g',linestyle=":")
plt.legend()
plt.show()
#plt.savefig("./myfig.png")
```
